```python
import pandas as pd

polestudios = pd.read_csv("Pole_Studio_Übersicht_S.csv")
polestudios1 = pd.read_csv("Pole_Studio_Übersicht_S_1.csv")
polestudios2 = pd.read_csv("Pole_Studio_Übersicht_S_2.csv")


workshops = pd.read_csv("Workshops_SW_E.csv")
workshops1 = pd.read_csv("Workshops_SW_E_1.csv")
workshops2 = pd.read_csv("Workshops_SW_E_2.csv")
```

```python
len(workshops)+len(workshops1)+len(workshops2)
```

```python
polestudios_final = pd.merge(polestudios, polestudios1, how='outer')
polestudios_final = pd.merge(polestudios_final, polestudios2, how='outer')
```

```python
polestudios_final
```

```python
workshops_final = pd.merge(workshops, workshops1, on=['Workshopname', 'URL_E'], how='inner')
```

```python
workshops_final = pd.merge(workshops, workshops1, how='outer')
workshops_final = pd.merge(workshops_final, workshops2, how='outer')
```

```python
workshops_final
```

```python
workshops_final.to_excel("workshops_final_E_SW.xlsx")
polestudios_final.to_excel("polestudios_final_S.xlsx")
```
